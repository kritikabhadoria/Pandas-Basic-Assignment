**Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.**

A DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. It is similar to a table in a relational database or an Excel spreadsheet. Each column in a DataFrame is a pandas Series. The main difference between DataFrame and Series is that DataFrame can hold multiple columns of data, while Series represents a single column.

Example:
```python
import pandas as pd

# Creating a Series
series = pd.Series([1, 2, 3, 4, 5])

# Creating a DataFrame from a dictionary
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)

print("Series:")
print(series)
print("\nDataFrame:")
print(df)
```

**Q7. Create a DataFrame using multiple Series. Explain with an example.**

```python
import pandas as pd

# Creating Series
name_series = pd.Series(['Alice', 'Bob', 'Charlie'])
age_series = pd.Series([25, 30, 35])
gender_series = pd.Series(['Female', 'Male', 'Male'])

# Creating DataFrame using multiple Series
df = pd.DataFrame({'Name': name_series, 'Age': age_series, 'Gender': gender_series})

print(df)
```

In this example, we create three Series: `name_series`, `age_series`, and `gender_series`. Then, we create a DataFrame `df` by passing these Series into a dictionary where the keys are the column names. Finally, we print the DataFrame which contains the data from the Series.
